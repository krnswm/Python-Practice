**LDA -** Linear Discriminant Analysis

```python
# It is a supervised machine learning technique
# It's main goal is to maximize class separability
```

example -> you have 2 types of fruits
    
     - type A -> "apple"
     - type B -> "Banana"

  "can i find a line so that when i project all the points onto that line , they will form two separate groups"

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_digits
```

```python
dataset = load_digits()
dataset
```

```python
df = pd.DataFrame(dataset.data, columns = dataset.feature_names) 
df
```

```python
 df['Target'] = dataset.target
```

```python
df
```

```python
# Toal number of unique values in the target column
df['Target'].unique()
```

```python
# Divide the data into x and y 
# ten perform the train test and split
```

```python
x = dataset.data
y = dataset.target
```

```python
x # independent features
```

```python
y # dependent feature
```

```python
x = df.drop('Target', axis = 1)
y = df['Target']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 3)
```

```python
x_train.shape
```

```python
y_train.shape
```

LDA works by finding linear combination of features that best separate the classes.

what if we have features with different scales or units -> we have to standardize the data then

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# we do not use fit on x_test because we don't to recalculate the mean and sd
```

```python
x_test.std()
```

```python
x_train.std()
```

```python
pd.DataFrame(x_train)
```

```python
# selecting the n components
```

* To find the number of components required for LDA, use the formula -

```python
# n_components = min(no. of features, no. of classes - 1)
```

```python
# number of features = 64
# number of classes = 10
# min(64, 10-1) => min(64,9)
# 9 is the minimum value so we will select 9 number of components
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis(n_components = 9)
```

```python
x_train = lda.fit_transform(x_train, y_train)
x_test = lda.transform(x_test)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model_pred = model.predict(x_test)
```

```python
from sklearn.metrics import *
```

```python
acc_score = accuracy_score(model_pred, y_test)
acc_score
```
