```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./cars.csv')
df
```

1. Create scatter plots for the following data, make sure all the plots appear in the same
plane with x-labels and y-labels evenly spaced.

a. Create scatter plots with between (mpg-hp, mpg-weight), and (hp-mpg,
hp-time-to-60) with respect to each brand

```python
fig = plt.subplots(2,2, figsize = (20,10))

#mpg-hp
plt.subplot(2,2,1)
sns.scatterplot(x = df['mpg'], y =df['hp'],data = df, hue =df['brand'])
plt.xlabel("mpg")
plt.ylabel("hp")
plt.title("mpg vs hp")

#mpg-weight
plt.subplot(2,2,2)
sns.scatterplot(x = df['mpg'], y = df['weightlbs'], data = df, hue = df['brand'])
plt.xlabel('mpg')
plt.ylabel('weight')
plt.title('mpg vs weight')

# hp-mpg
plt.subplot(2,2,3)
sns.scatterplot(x = df['hp'], y =df['mpg'],data = df, hue =df['brand'])
plt.xlabel("hp")
plt.ylabel("mpg")
plt.title(" hp vsmpg")

#hp-time-to-60

plt.subplot(2,2,4)
sns.scatterplot(x = df['hp'], y =df['time-to-60'],data = df, hue =df['brand'])
plt.xlabel("hp")
plt.ylabel("time-to-60")
plt.title(" hp vs time-to-60")

plt.show()
```

2. Create a histogram for the following:

a. Create histograms for hp, mpg, time-to-60, weight lbs, cubic inches, year, with
respect to brand.


```python
fig=plt.subplots(3,2,figsize=(20,20))

plt.subplot(3,2,1)
sns.histplot(data=df,x=df['hp'],hue=df['brand'])
plt.xlabel('HP')
plt.ylabel('Frequency')
plt.title('Histogram of HP')

plt.subplot(3,2,2)
sns.histplot(data=df,x=df['mpg'],hue=df['brand'])
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.title('Histogram of MPG')

plt.subplot(3,2,3)
sns.histplot(data=df,x=df['time-to-60'],hue=df['brand'])
plt.xlabel('Time-to-60')
plt.ylabel('Frequency')
plt.title('Histogram of Time')

plt.subplot(3,2,4)
sns.histplot(data=df,x=df['weightlbs'],hue=df['brand'])
plt.xlabel('WeightLBS')
plt.ylabel('Frequency')
plt.title('Histogram of WeightLBS')

plt.subplot(3,2,5)
sns.histplot(data=df,x=df['cubicinches'],hue=df['brand'])
plt.xlabel('Cubic Inches')
plt.ylabel('Frequency')
plt.title('Histogram of Cubic Inches')

plt.subplot(3,2,6)
sns.histplot(data=df,x=df['year'],hue=df['brand'])
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Histogram of Year')

plt.show()
```
