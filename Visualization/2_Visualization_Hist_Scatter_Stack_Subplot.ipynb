### Scatter Plot

```python
# It is used between two continuous columns
# It is used to understand relation beteen two continuous columns
```

```python
# Mileage vs Weight
# If weight is increasing then mileage will reduce
```

```python
plt.scatter(df1['mpg'], df1['weight'])
```

```python
plt.scatter(df1['weight'], df1['displacement'])
plt.xlabel('Weight')
plt.ylabel('Disp.')
plt.title('Scatter Plot - Weight v/s Displacement')
plt.show()
```

```python
plt.subplot(121)
plt.scatter(df1['mpg'], df1['weight'])

plt.xlabel('MPG')
plt.ylabel('Weight.')
plt.title('Scatter Plot - MPG v/s Weight')

plt.subplot(122)
plt.scatter(df1['mpg'], df1['displacement'])

plt.xlabel('MPG')
plt.ylabel('Disp.')
plt.title('Scatter Plot - MPG v/s Displacement')

plt.subplots_adjust(wspace = 0.5, hspace = 0.5)
plt.show()
```

```python
A = [21,22,23,24,25]
B = [24,25,22,24,21]
C = [1,2,3,4,5]

plt.subplot(221)
plt.scatter(C,A, marker = 'D', color = 'Black')

plt.subplot(222)
plt.plot(C,A, linestyle = 'dashed', color = 'Red', marker = '+') # Line PLot

plt.subplot(223)
plt.scatter(C,B)

plt.subplot(224)
plt.stackplot(C,B)  #Stack Plot

plt.show()
```
