```python
# 3) Check for outliers

for i in df.columns:
    if df[i].dtypes == 'int64' or df[i].dtypes == 'float64':
        plt.boxplot(df[i])
        plt.xlabel(i)
        plt.show()
```

```python
# How we remove the outliers

for i in df.columns:
    if df[i].dtypes != 'object':
        Q1 = df[i].quantile(0.25)
        Q3 = df[i].quantile(0.75)
        IQR = Q3 - Q1
        UB = Q3 + 1.5 * IQR
        LB = Q1 - 1.5 * IQR

        #Put a condition that if any data is outside the upper range and lower range that is my outlier
        # Consider data in this range

        df = df[(df[i] >= LB) & (df[i] <= UB)]
```

```python
for i in df.columns:
    if df[i].dtypes == 'int64' or df[i].dtypes == 'float64':
        plt.boxplot(df[i])
        plt.xlabel(i)
        plt.show()
```
