```python
# Applying a function to a DataFrame
```

```python
df1.head(2)
```

```python
res = []
for i in df1['acceleration']:
    if i%2 == 0:
        res.append('Even')
    else:
        res.append('Odd')
```

```python
df1['col1'] = res
df1.head()
```

```python
def even_odd(num):
    if num%2 == 0:
        return 'Even'
    else:
        return 'Odd'
```

```python
even_odd(3)
```

```python
even_odd(4)
```

```python
df1['col2'] = df1['acceleration'].apply(even_odd)
df1.head()
```

```python
# Lambda
# Anonymous Function / One time Execution

df1['col3'] = df1['acceleration'].apply(lambda num: 'Even' if num%2 == 0 else 'Odd')
df1.head()
```

```python
df1['col4'] = df1['acceleration'].apply(lambda num: num**2)
df1.head(2)
```
