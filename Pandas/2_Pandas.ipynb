```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'id': [1,2,3,4,5], 'stat_marks': [94,98,99,92,90]})
df2 = pd.DataFrame({'id': [3,4,5,6,7], 'ml_marks': [90,92,96,89,94]})
df2
```

### Merge

```python
# SQL join == pandas Join
```

```python
# It will return results of students who appeared in both exams
df1.merge(df2, on = 'id', how = 'inner')
```

```python
df1.merge(df2, on = 'id', how = 'left')  # 1, 2 didn't appeared on ML exam --> Null == NaN
# NaN --> Null/Blank/Empty
```

```python
df1.merge(df2, on = 'id', how = 'right')
```

```python
df1.merge(df2, on = 'id', how = 'outer')
```

```python
df1.merge(df2, how = 'cross')
```

 ### Concat

```python
# SQL union = Pandas Concat
```

```python
df1 = pd.DataFrame({'id': [1,2,3,4,5], 'stat_marks': [94,98,99,92,90]})  # master data base (Jan - Aug)
df2 = pd.DataFrame({'id': [6,7,8,9,10], 'stat_marks': [90,92,96,89,94]})  # append latest data to master database (sept)
df2
```

```python
pd.concat([df1,df2])
```

```python
df3 = pd.concat([df1,df2])
df3

df3.reset_index() # by default stores old index in a column adn resets the index and does not change the original index

df3.reset_index(drop = True) # drop = True --> will not save old index column

# df3 = df3.reset_index(drop = True) # commiting changes to df3
# or what we can do is,
df3.reset_index(drop = True, inplace = True) # inplace = True --> will commit changes to df3

df3
```
