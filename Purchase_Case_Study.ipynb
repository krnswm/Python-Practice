It was observed that the average purchase made by men of the age 18-25 was 10000, Is it still the same

```python
new_data = df[(df['Age'] == 1) & df['Gender'] == 1]
new_data
```

```python
sample = new_data.sample(3600, random_state = 50)
sample
```

```python
# null - the average purchase is same
# alt - the average puchase is not the same
```

```python
# import any test
from scipy.stats import ttest_1samp
a_mean = 10000
```

```python
t_stat, p_val = ttest_1samp(sample['Purchase'], a_mean)
print(p_val)
```

```python
# as te p-value < 0.05 so we are rejecting the null hypo and considering that the average purchase is not same
```

Is the average purchase made by men and women of the age 18-25 same

```python
#null - avg purchase is equal
#alt - avg purchase is not equal
```

```python
dfm = df.loc[(df['Gender'] == 1) & (df['Age'] == 1)]
dfw = df.loc[(df['Gender'] == 0) & (df['Age'] == 1)]
```

```python
dfm
```

```python
dfw
```

```python
dfms = dfm.sample(1000, random_state = 1)
dfws = dfw.sample(1000, random_state = 7)
```

```python
m_mean = dfms['Purchase'].mean()
m_mean
```

```python
w_mean = dfws['Purchase'].mean()
w_mean
```

```python
from scipy.stats import ttest_1samp
```

```python
t_stat, p_val = ttest_1samp(dfms['Purchase'], m_mean)
print(p_val)
```

```python
t_stat, p_val = ttest_1samp(dfws['Purchase'], w_mean)
print(p_val)
```
