## What is VIF (Value Inflation Factor)
- Multicolinearity occurs when two or more independent columns in a dataset are highly correlated with each other
- VIF is used to detect multicollienar columns
- If VIF score is < 6 means that columns are good to go
- If VIF score is > 6 means columns are multicolinear columns

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Create a list of all numerical columns
col_list=[]
for i in df.columns:
    if (df[i].dtype != 'object') & (i != 'charges'):
        col_list.append(i)

# Apply VIF on the columns present inside the list
X = df[col_list]  # Storing all the columns in the variable X
vif_data = pd.DataFrame()
vif_data['Column_Name'] = X.columns
vif_data['VIF_Score'] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]
vif_data
```

## Model Building

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
# Train the Model
model.fit(x_train, y_train)
```
