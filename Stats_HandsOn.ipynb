```python
# IQR = Q3 -Q1
# LB = Q1 - 1.5 * Q3
# UB = Q1 + 1.5 * Q3
```

```python
Q1 = df['Population in thousands (2017)'].quantile(0.25)
Q1
```

```python
Q3 = df['Population in thousands (2017)'].quantile(0.75)
Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
LB = Q1 - 1.5 * Q3
print(LB)
UB = Q1 + 1.5 * Q3
print(UB)
```

```python
df = df[(df['Population in thousands (2017)'] >= LB) & (df['Population in thousands (2017)'] <= UB)]
df
```
