```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)
```

```python
x1 = 3
x2 = 2
y_actual = 17
```

```python
np.random.seed(42)
w1 = np.random.randint(1, 10)
w2 = np.random.randint(1, 10)

learning_rate = 0.01

epochs = 20
errors = []
predictions = []

print("Initial Weights:")
print("w1 =", w1)
print("w2 =", w2)
```

```python
for epoch in range(epochs):

    # Feed Forward
    y_pred = x1 * w1 + x2 * w2
    
    # Error Calculation (Squared Error)
    error = (y_actual - y_pred) ** 2
    
    errors.append(error)
    predictions.append(y_pred)
    
    # Backpropagation (Gradient Calculation)
    dE_dw1 = -2 * x1 * (y_actual - y_pred)
    dE_dw2 = -2 * x2 * (y_actual - y_pred)
    
    # Update Weights
    w1 = w1 - learning_rate * dE_dw1
    w2 = w2 - learning_rate * dE_dw2

print("\nFinal Weights:")
print("w1 =", w1)
print("w2 =", w2)
```

```python
# Plot Error
plt.figure()
plt.plot(errors)
plt.title("Error vs Epochs")
plt.xlabel("Epoch")
plt.ylabel("Squared Error")
plt.show()

# Plot Prediction
plt.figure()
plt.plot(predictions)
plt.axhline(y=17)
plt.title("Prediction vs Epochs")
plt.xlabel("Epoch")
plt.ylabel("Predicted Value")
plt.show()
```
