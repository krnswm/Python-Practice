```python
# Dummy Dataset
import numpy as np

#Random Dataset
X_train=np.random.random((1000,20))
y_train=np.random.random((1000,1))


X_test=np.random.random((100,20))
y_test=np.random.random((100,1))
```

# Creating a Model

## Layers


*   Dense Layer -> A layer of fully connected neurons
*   Dropout
*   Conv1D, conv2D,
*   LSTM/GRU



```python
# Imports
from keras.models import Sequential
from keras.layers import Dense

from keras import optimizers
from keras import metrics
```

```python
# define the structure of the model
model=Sequential()

# Input Layer / Hidden Layer
model.add(Dense(10,activation='relu',input_dim=X_train.shape[1]))

# Second Layer
model.add(Dense(10,activation='relu'))

# Output Layer
model.add(Dense(1,activation='relu'))
model.summary()
```

```python
# Configure the model for learning/training
# Loss, metrics, optimizer

model.compile(loss='mse',
              optimizer='adamax',
              metrics=['accuracy'])
```

```python
# 1000- Size, BS:500, 30 epochs
# (1000/500)*30 =60
```

```python
# Training the model
model.fit(X_train,y_train,batch_size=1,epochs=30,validation_data=(X_test,y_test))
```
